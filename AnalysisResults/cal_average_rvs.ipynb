```python
import json
from collections import defaultdict
```

```python
pathin = '../SimulationResults/'
model = 'NonlinearHC_SIR'
index = 'range'  # 'length'
it = 'ini'
net = [1]
para = 1
expr = 2
# file = f'{path}{model}/{net}_para{para}_expr{expr}_{index}.json'
```

```python
for i in net:
    file = f'{pathin}{model}/{i}_para{para}_expr{expr}_{index}.json'
    with open(f'{file}', 'r') as f:
        result = json.load(f)
    nodeAv = defaultdict(float)
    for item in result:  
        name = item.split('_')
        keyname = name[3][3:]
        nodeAv[keyname] = sum(result[item])/len(result[item])

    avNode = defaultdict(list)
    for node in nodeAv:
        avNode[nodeAv[node]].append(int(node))
    keysLstStr = avNode.keys()
    keysLst = [float(i) for i in keysLstStr]  # if duplicates
    keysSet = set(keysLst)
    duplicates = []
    if len(keysLst) == len(keysSet):
        print(f"net{i} no duplicates")
    else:
        for element in keysLst:
            if keysLst.count(element) > 1 and element not in duplicates:
                duplicates.append(element)
        print(f"net{i} duplicates={duplicates}, further processing needed!")

    # with open(f'{model}_avg/avg_{i}_para{para}_expr{expr}_{index}.json', 'w') as g:
    #     json.dump(nodeAv, g)
    # with open(f'{model}_avg/avg_{i}_para{para}_expr{expr}_{index}_rvs.json', 'w') as f:
    #     json.dump(avNode, f)
```
